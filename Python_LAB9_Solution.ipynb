```python
#create table for May
df_rental_may = pd.read_sql_query(
"""
select
	r.customer_id
    ,c.first_name
    ,c.last_name
	,count(*) as num_rentals_may
from rentals_may r
	inner join customer c on r.customer_id=c.customer_id
group by r.customer_id,c.first_name, c.last_name
order by num_rentals_may desc;;
""", engine
)
df_rental_may.head()
```

```python
#create table for June
df_rental_june = pd.read_sql_query(
"""
select
	r.customer_id
    ,c.first_name
    ,c.last_name
	,count(*) as num_rentals_june
from rentals_june r
	inner join customer c on r.customer_id=c.customer_id
group by r.customer_id,c.first_name, c.last_name
order by num_rentals_june desc;;
""", engine
)
df_rental_june.head()
```
